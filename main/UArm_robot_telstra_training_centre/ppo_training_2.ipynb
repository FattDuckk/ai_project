```python
from stable_baselines3 import PPO
from env import UArmEnv
from stable_baselines3.common.callbacks import BaseCallback
import numpy as np

# Optional logging callback
class RewardAndGoalLoggingCallback(BaseCallback):
    def __init__(self, log_interval=20000, verbose=0):
        super().__init__(verbose)
        self.log_interval = log_interval
        self.cumulative_reward = 0
        self.step_counter = 0

    def _on_step(self):
        reward = self.locals["rewards"][0]
        self.cumulative_reward += reward
        self.step_counter += 1

        if self.step_counter % self.log_interval == 0:
            goal_pos = self.training_env.get_attr("goal_pos")[0]
            ef_pos = self.training_env.get_attr("robot")[0].get_joint_obs()["ee_pos"]
            print(f"üìà Step {self.num_timesteps} | Reward: {round(self.cumulative_reward, 4)} | üéØ Goal: {np.round(goal_pos, 3)} | ü§ñ EE: {np.round(ef_pos, 3)}")
            self.cumulative_reward = 0

        return True

# Train model
env = UArmEnv(render=False)
model = PPO("MlpPolicy", env, verbose=1)
callback = RewardAndGoalLoggingCallback(log_interval=20000)
model.learn(total_timesteps=1_000_000, callback=callback)
model.save("ppo_uarm_final")
env.close()
```
